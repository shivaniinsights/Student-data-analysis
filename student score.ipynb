Drop unnamed column

```python
df=df.drop(f"ID:0",axis=1)
df.head()
```

change weekly study hours coolumn

```python
df["WklyStudyHours"] =df["WklyStudyHours"].str.replace("05-oct","5-10")
df
```

gender distribution

```python
plt.figure(figsize=(6,3))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analysed that the number of females in the data 
#is more than the numberof males
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
gb
```

```python
sns.heatmap(gb,annot=True)
plt.show()
```

from the above chart we have concluded that the education of the parents have a good impact

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
gb1
```

```python
sns.heatmap(gb1,annot=True)
plt.title("Relatiionship between parent's Marital status and student's score")
plt.show()
```

The above chart we have concluded that there is no/negligible impact on the student's score due to their parent's  marital status

Distribution of Ethnic groups

```python
groupA=df.loc[(df['EthnicGroup']== "group A")].count()
groupB=df.loc[(df['EthnicGroup']== "group B")].count()
groupC=df.loc[(df['EthnicGroup']== "group C")].count()
groupD=df.loc[(df['EthnicGroup']== "group D")].count()
groupE=df.loc[(df['EthnicGroup']== "group E")].count()

mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie([mlist])
plt.show()
```

```python
plt.pie(mlist, labels=["Group A", "Group B", "Group C", "Group D", "Group E"], autopct='%1.1f%%')
plt.title("Distribution of Ethnic Groups")
plt.show()
```
